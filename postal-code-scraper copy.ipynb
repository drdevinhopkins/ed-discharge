```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

def get_CIUSSS_by_postalcode(postalcode):
    url = "https://sante.gouv.qc.ca/repertoire-ressources/resultats-recherche/?theme=votre-cisss-ciusss&ch_code={}&bt_rechCode=".format(postalcode)
    page = urlopen(url)
    html = page.read().decode("utf-8")
    soup = BeautifulSoup(html, "html.parser")
    ciusss = str(soup.find('span', {'class' : 'liste__resultats-lien'})).split('>')[2].split('<')[0]
    return ciusss
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd


DRIVER_PATH = '/path/to/chromedriver'
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
driver.get('https://m34.pub.msss.rtss.qc.ca/rechCodePostal.asp')
time.sleep(1)
choose_ciusss = driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/table/tbody/tr[4]/td/div/input[3]')
choose_ciusss.click()
postal_code_input = driver.find_element(By.ID, 'txtCodePostal')
postal_code_input.clear()
postal_code_input.send_keys('H4A')
postal_code_input.send_keys(Keys.RETURN)
time.sleep(1)
```

```python
html = driver.page_source
df = pd.read_html(html)
df = df[3]
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df
```

```python
path = '/html/body/table/tbody/tr[5]/td/form/table[2]/tbody/tr'

len(driver.find_elements(By.XPATH,path))
```

```python

# Obtain the number of rows in body
rows = 1+len(driver.find_elements(By.XPATH,
    '/html/body/table/tbody/tr[5]/td/form/table[2]/tbody/tr'
))
  
# Obtain the number of columns in table
cols = len(driver.find_elements(By.XPATH,
    '/html/body/table/tbody/tr[5]/td/form/table[2]/tbody/tr[2]/td'))
  
# Print rows and columns
print(rows)
print(cols)
```

```python
# Printing the data of the table
for r in range(2, rows+1):
    for p in range(1, cols+1):
        
        # obtaining the text from each column of the table
        value = driver.find_element(By.XPATH,
            "/html/body/table/tbody/tr[5]/td/form/table[2]/tbody/tr["+str(r)+"]/td["+str(p)+"]").text
        print(value, end='       ')
    print()
```
