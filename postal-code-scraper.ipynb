```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd


# DRIVER_PATH = '/path/to/chromedriver'
# driver = webdriver.Chrome(executable_path=DRIVER_PATH)

def get_ciusss_by_postalcode(postalcode):
        
    output = pd.DataFrame()

    driver.get('https://m34.pub.msss.rtss.qc.ca/rechCodePostal.asp')
    time.sleep(1)

    choose_ciusss = driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/table/tbody/tr[4]/td/div/input[3]')
    choose_ciusss.click()
    postal_code_input = driver.find_element(By.ID, 'txtCodePostal')
    postal_code_input.clear()
    postal_code_input.send_keys(postalcode)
    postal_code_input.send_keys(Keys.RETURN)
    time.sleep(1)

    html = driver.page_source
    df = pd.read_html(html)
    df = df[3]
    new_header = df.iloc[0] #grab the first row for the header
    df = df[1:] #take the data less the header row
    df.columns = new_header #set the header row as the df header
    output = pd.concat([output, df], ignore_index=True)

    cur_page = int(driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/span/b[1]').text)
    num_pages = int(driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/span/b[2]').text)

    while cur_page < num_pages:
        next_page = driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/a[2]')
        next_page.click()
        time.sleep(1)

        html = driver.page_source
        df = pd.read_html(html)
        df = df[3]
        new_header = df.iloc[0] #grab the first row for the header
        df = df[1:] #take the data less the header row
        df.columns = new_header #set the header row as the df header
        output = pd.concat([output, df], ignore_index=True)

        cur_page = int(driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/span/b[1]').text)
        num_pages = int(driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/span/b[2]').text)

        # print(len(output))
    output.to_csv('scraped-csvs/'+postalcode+'.csv', index=False)
    time.sleep(5)
    return output
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd


# DRIVER_PATH = '/path/to/chromedriver'
# driver = webdriver.Chrome(executable_path=DRIVER_PATH)

def get_clsc_by_postalcode(postalcode):
        
    output = pd.DataFrame()

    driver.get('https://m34.pub.msss.rtss.qc.ca/rechCodePostal.asp')
    time.sleep(1)

    choose_clsc = driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/table/tbody/tr[4]/td/div/input[2]')
    choose_clsc.click()
    postal_code_input = driver.find_element(By.ID, 'txtCodePostal')
    postal_code_input.clear()
    postal_code_input.send_keys(postalcode)
    postal_code_input.send_keys(Keys.RETURN)
    time.sleep(1)

    html = driver.page_source
    df = pd.read_html(html)
    df = df[3]
    new_header = df.iloc[0] #grab the first row for the header
    df = df[1:] #take the data less the header row
    df.columns = new_header #set the header row as the df header
    output = pd.concat([output, df], ignore_index=True)

    cur_page = int(driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/span/b[1]').text)
    num_pages = int(driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/span/b[2]').text)

    while cur_page < num_pages:
        next_page = driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/a[2]')
        next_page.click()
        time.sleep(1)

        html = driver.page_source
        df = pd.read_html(html)
        df = df[3]
        new_header = df.iloc[0] #grab the first row for the header
        df = df[1:] #take the data less the header row
        df.columns = new_header #set the header row as the df header
        output = pd.concat([output, df], ignore_index=True)

        cur_page = int(driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/span/b[1]').text)
        num_pages = int(driver.find_element(By.XPATH, '/html/body/table/tbody/tr[5]/td/form/span/b[2]').text)

        # print(len(output))
    output.to_csv('clsc-scraped-csvs/'+postalcode+'.csv', index=False)
    time.sleep(1)
    return output
```

```python
fsa_df = pd.read_csv('quebec-fsas.csv')
fsas = fsa_df.fsa.tolist()
```

```python
final = pd.DataFrame()
i = 0
for postalcode in fsas:
    i = i+1
    print(postalcode, i, 'of', len(fsas))
    output = get_clsc_by_postalcode(postalcode)
    # print(output)
    final = pd.concat([final, output[['Code postal', 'Nom CLSC(Territoire)']]], ignore_index=True)
final.to_csv('clsc-final.csv', index=False)
```
